## Gather

```python
import pandas as pd
import os
import requests 
import json
import tweepy
from timeit import default_timer as timer
```

```python
# The first given file
df_enhanced = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
# using request library to save data from the image_predictions.tsv file
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
response = requests.get(url)
fileName = 'image-predictions.tsv'
with open(fileName, mode = 'wb') as file:
        file.write(response.content)
```

```python
df_predictions = pd.read_csv(fileName, sep='\t')
```

```python
# using Twitter Api (Tweepy) to get other needed info.

# Here are the required keys to access tweepy (found on twitter developer account)
consumer_key = '--------------'
consumer_secret = '-------------'
access_token = '---------------'
access_token_secret = '----------------'

# Setup access to API
def connect_to_twitter_OAuth():
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    
    # wait on rate limit is used to give notification when data retreiving rate limit in reached 
    # and continue again after a while
    api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
    return api

# Create API object
api = connect_to_twitter_OAuth()
```

```python
# retreiving data from twiiter upon the providded tweet ids from df_enhanced
errors = {}
progress = 0

start = timer()

with open('tweet_json.txt', 'w') as outfile:
    for t_id in df_enhanced['tweet_id']:
        progress+=1
        print('Progress: {}'.format(progress))

        try:
            tweet = api.get_status(t_id, tweet_mode='extended')
            print('Successfull task')
            json.dump(tweet._json, outfile)
            outfile.write('\n') 

        except tweepy.TweepError as e:
            print('Error found: {}'.format(e))
            errors[t_id] = e
            pass
        
end = timer()
print(end - start)
print(errors)
```

```python
#checking the api columns
df_api = pd.read_json('tweet_json.txt', lines=True)
print(df_api.columns)
```

# Clean

```python
# copy of each dataframe
new_enhanced = df_enhanced.copy()
new_predictions = df_predictions.copy()
new_api = df_api.copy()
```

## Quality issues

**1. In df_enhanced, remove Wrong puppies names as (a, an, the, very, all, ...)**

```python
#code
for i, puppy_name in enumerate(new_enhanced['name']):
    if(puppy_name == 'a' or puppy_name == 'an' or puppy_name == 'the' or puppy_name == 'very' or puppy_name == 'all' or puppy_name == 'this'):
        new_enhanced['name'][i] = 'None'

#test
new_enhanced.query('name == "a"')

#Now, The result shows no rows where name is 'a', I also checked the other wrong names and they has been removed too.
```

**2. In df_enhanced, at index 313, the denumerator equals zero so it will be removed**

```python
#code
new_enhanced.drop(new_enhanced.index[313], inplace = True)

#test
new_enhanced.iloc[313]['rating_denominator']                                                          


#Now, at index 313, the denominator is not zero which means the row has been deleted
```

**3. In df_enhanced, convert timestamp to datetime to make the columns: year, month and day**

```python
#code
new_enhanced['Day'] = pd.to_datetime(new_enhanced['timestamp']).dt.day
new_enhanced['Month'] = pd.to_datetime(new_enhanced['timestamp']).dt.month
new_enhanced['Year'] = pd.to_datetime(new_enhanced['timestamp']).dt.year

#test
new_enhanced[['Day' ,'Month', 'Year']]
```

**4. In df_enhanced, Create one column for rating, equals num/denom to remove redundancy in denomerator column (where most of values are 10).**

```python
#code
new_enhanced['rating_ratio'] = new_enhanced['rating_numerator']/new_enhanced['rating_denominator']

#test
new_enhanced[['rating_ratio']]
```

**5. In df_predictions file, remove duplicated jpg_url rows**

```python
#code 
new_predictions.drop_duplicates(subset=['jpg_url'], inplace = True)

#test
new_predictions.jpg_url.duplicated().sum()
```

**6. In df_predictions file, remove rows where all predictions are False as they are not needed for the analysis**

```python
#code
wrong_predictions = new_predictions[['p1_dog', 'p2_dog', 'p3_dog']].query('p1_dog == False and p2_dog == False and p3_dog == False').index

# wrong_predictions is a list of all the indexes where the row should be deleted, so I will make new column 'Marked'
# that holds 'True' value at these indexes so that I can remove these rows, finally I will drop the Marked column again
new_predictions['Marked'] = False
new_predictions.at[wrong_predictions, 'Marked'] = True 

#removing rows with True value at 'Marked' column
new_predictions = new_predictions[new_predictions['Marked'] != True]

#resetting index because removed indexes are now missing, and dropping the unneeded 'marked' column
new_predictions.reset_index()
new_predictions.drop(['Marked'], axis = 1, inplace = True)

#test
new_predictions[['p1_dog', 'p2_dog', 'p3_dog']].query('p1_dog == False and p2_dog == False and p3_dog == False')
```

**7. In df_predictions file, rename some columns to be more descriptive**

```python
#code
new_predictions.rename(columns = {'p1' : 'Prediction_1', 'p2' : 'Prediction_2', 'p3' : 'Prediction_3', 
                                  'p1_conf' : 'confidence_1', 'p2_conf' : 'confidence_2', 'p3_conf' : 'confidence_3',
                                  'p1_dog' : 'isDogType_1', 'p2_dog' : 'isDogType_2', 'p3_dog' : 'isDogType_3'}, 
                       inplace = True)

#test
new_predictions.head()
```

**8. In df_api, delete retweets rows**

```python
#code

#check if column 'in_reply_to_status_id' is not null, if so, then it's a retweet then should be removed
new_api = new_api[new_api.in_reply_to_status_id.isnull() != False]

#test: if all results of 'isnull()' are True
new_api.in_reply_to_status_id.isnull().value_counts()
```

```python
#now there is no need for the 'in_reply_to_status_id' column as it is all "NaN"
new_api.drop(['in_reply_to_status_id'], axis = 1, inplace = True)
```

## Tidiness issues

**1. In df_enhanced, turn the last four columns into one column with 'category' type and values: doggo, floofer, pupper and puppo**

```python
#code
new_enhanced['dog_stage'] = 'None'

#for each of the columns doggo, floofer, pupper and puppo, if there is a value, insert it to the 'dog_stage' column
new_enhanced.at[new_enhanced['doggo'] == 'doggo', 'dog_stage'] = 'doggo'
new_enhanced.at[new_enhanced['floofer'] == 'floofer', 'dog_stage'] = 'floofer'
new_enhanced.at[new_enhanced['pupper'] == 'pupper', 'dog_stage'] = 'pupper'
new_enhanced.at[new_enhanced['puppo'] == 'puppo', 'dog_stage'] = 'puppo'

#test
new_enhanced.info()
```

```python
# now, there is no need for the four columns: doggo, floofer, pupper and puppo
new_enhanced.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis = 1, inplace = True)
new_enhanced.info()
```

```python
new_api.info()
```

**2. For all dataframes, delete unnecessary columns**


```python
#code

#new dataframes with only the needed columns
final_enhanced = new_enhanced.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 
                                    'timestamp', 'retweeted_status_id', 'retweeted_status_user_id', 
                                    'retweeted_status_timestamp', 'expanded_urls', 'rating_numerator', 
                                    'rating_denominator', 'source'], axis = 1)

final_api = new_api.drop(['truncated', 'entities', 'extended_entities', 'in_reply_to_status_id_str', 
                          'in_reply_to_user_id', 'in_reply_to_user_id_str', 'in_reply_to_screen_name', 
                          'contributors', 'is_quote_status', 'possibly_sensitive', 'possibly_sensitive_appealable', 
                          'quoted_status_id', 'quoted_status_id_str', 'quoted_status_permalink', 'quoted_status', 
                          'created_at', 'id_str', 'user', 'coordinates', 'place', 'geo', 'full_text', 'source',
                          'lang', 'display_text_range', 'retweeted_status', 'favorited', 'retweeted'], axis = 1)

#test
print(final_enhanced.columns)
print()
print(final_api.columns)
```

**3. Merge dataframes**

```python
#code
df_final = pd.merge(final_enhanced, final_api.rename(columns= {'id':'tweet_id'}), on = 'tweet_id', how = 'left')
df_final = pd.merge(df_final, new_predictions, on = 'tweet_id', how = 'left')

#test
df_final.sample(10)
```
